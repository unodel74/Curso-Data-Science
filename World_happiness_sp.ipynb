```python
# Agrupo por regiones y muestro el valor máximo en cada grupo. Y lo ordeno de mayor a menor
data_max_average = database.groupby(['region'])['Happiness'].max()
data_max_average.sort_values(ascending=False)
```

**Outliers**

```python
# Función que cuenta el número de outliers
def count_outliers(df: pd.DataFrame, columna:str) -> int:
  Q1 = np.percentile(df[columna], 25)
  Q2 = np.percentile(df[columna], 50)
  Q3 = np.percentile(df[columna], 75)

  IQR = Q3 - Q1

  iqr_lower_limit = Q1 - 1.5 * IQR
  iqr_upper_limit = Q3 + 1.5 * IQR

  outliers_iqr = df[columna][(df[columna] < iqr_lower_limit) | (df[columna] > iqr_upper_limit)]
  return len(outliers_iqr)
```

```python
# Compruebo el número de outliers

count_outliers(database, "Happiness")
```

**Estandarización**

```python
# Voy a reescalar los datos, para una mayor facilidad de interpretación.

import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

data_scaled = database.copy()
zscore_scaler = StandardScaler()
df_standarized = zscore_scaler.fit_transform(data_scaled[['Happiness']])

plt.figure(figsize=(6, 4))
sns.histplot(df_standarized, bins=20, color='skyblue')
plt.title('Distribución de la Felicidad')
plt.xlabel('Felicidad')
plt.ylabel('Nivel')
plt.grid(True)
plt.show()
```

**Escalado Robusto**

```python
# Escalado Robusto: este escalador elimina la mediana y escala los datos según el rango intercuartílico. (IQR: rango entre el primer cuartil y el tercer cuartil). Es ideal para datos con outliers, ya que el IQR es robusto a ellos.


data_scaled = database.copy()

robust_scaler = RobustScaler()
df_robust = robust_scaler.fit_transform(data_scaled[['Happiness']])

plt.figure(figsize=(6, 4))
sns.histplot(df_robust, bins=20, color='skyblue')
plt.title('Distribución de la Felicidad')
plt.xlabel('Felicidad')
plt.ylabel('Nivel')
plt.grid(True)
plt.show()
```

# **Conclusiones**

**-** La primera conclusión es que, en general, es cierto que **el dinero da la felicidad**. Vemos que los países con rentas altas, son los que tienen mayores índices de felicidad.
**Aunque encontramos casos de países con rentas bajas que tienen índices de felicidad altos**, como es el caso de Costa Rica y otros países de Centroamérica y Sudamérica.

**-**  **La  felicidad es también muy proporcional a la esperanza de vida**. Los países con más esperanza de vida tienen valores altos de felicidad. Aunque la esperanza de vida vemos que también está muy relacionada con el nivel de ingresos.

**-**  Otra de las conclusiones que saco es que **en los paises con mayores rentas, la población más feliz es la de más edad**. Mientras que **en los países con menos rentas, son los jovenes los más felices**.

**-**  **La corrupción es un dato inversamente proporcional a todos los demás factores**. Sobre todo en cuanto a felicidad, ingresos y acceso a internet.

**-** Respecto al acceso a internet, tengo mis dudas. Por un lado, en la matriz de correlación, vemos que es proporcional a la felicidad. Pero, al igual que pasa con los ingresos, hay países en los que el acceso a internet es bajo y la felicidad es alta.

Por lo que, yo creo, que **el acceso a internet tiene que ver más con el nivel de ingresos, que con la felicidad**. Como se ve en el gráfico de dispersión entre esas dos variables.
